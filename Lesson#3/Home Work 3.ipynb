Напишите код на Python, реализующий построение графиков:
- окружности
- эллипса

```python
# ### -- График окружности --
# ### -----------------------
plt.figure(figsize=(5, 5))
x = np.linspace(-1.0, 1.0, 100)
y = np.linspace(-1.0, 1.0, 100)
x_axis = np.zeros(100)
y_axis = np.zeros(100)
X, Y = np.meshgrid(x, y)
F = X ** 2 + Y ** 2 - 0.5
plt.contour(X, Y, F, [0])
plt.plot(x_axis, y, 'c-.')
plt.plot(x, y_axis, 'c-.')
plt.grid(True, linestyle='-', color='0.75')
plt.gca().set_aspect('equal')
plt.show()
```

```python
# ### -- График элипса --
# ### -------------------
plt.figure(figsize=(5, 5))
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
x_axis = np.zeros(100)
y_axis = np.zeros(100)
a = 6 ** 2
b = 4 ** 2
X, Y = np.meshgrid(x, y)
F = (X ** 2 / a) + (Y ** 2 / b) - 0.5
plt.contour(X, Y, F, [0])
plt.plot(x_axis, y, 'c-.')
plt.plot(x, y_axis, 'c-.')
plt.grid(True, linestyle='-', color='0.75')
plt.gca().set_aspect('equal')
plt.show()
```

```python
# ### -- График гиперболы --
# ### ----------------------
plt.figure(figsize=(5, 5))
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
x_axis = np.zeros(100)
y_axis = np.zeros(100)
a = 3 ** 2
b = 5 ** 2
X, Y = np.meshgrid(x, y)
F = (X ** 2 / a) - (Y ** 2 / b) - 0.5
plt.contour(X, Y, F, [0])
plt.plot(x_axis, y, 'c-.')
plt.plot(x, y_axis, 'c-.')
plt.grid(True, linestyle='-', color='0.75')
plt.gca().set_aspect('equal')
plt.show()
```

3.1 Напишите код, который будет переводить полярные координаты в декартовы

```python
def polar2decart2D(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

def decart2polar2D(x, y):
    r = sqrt(x ** 2 + y ** 2)
    phi = np.arctan2(y, x)
    return r, phi
```

```python
print(f'{polar2decart2D(10, 45)}')
print(f'{decart2polar2D(5, 5)}')
```

4.1 Решить систему уравнений

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-5, 5, 500)
plt.plot(x, ((np.exp(x) - 1)/x + 1))
plt.plot(x, x**2 - 1, color='green')
plt.grid(True, linestyle='-', color='0.75')
plt.show()
```

```python
def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)

x1, y1 = fsolve(f, (-2, 0))
x2, y2 = fsolve(f, (3, 6))
x3, y3 = fsolve(f, (4, 17))
print(x1, y1)
print(x2, y2)
print(x3, y3)
```

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-5, 5, 500)
plt.plot(x, ((np.exp(x) - 1)/x + 1))
plt.plot(x, x**2 - 1, color='green')
plt.grid(True, linestyle='-', color='0.75')
plt.scatter(x1, y1, c = 'deeppink')
plt.scatter(x2, y2, c = 'deeppink')
plt.scatter(x3, y3, c = 'deeppink')
plt.show()
```

4.2 Решить систему уравнений и неравенств

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-4, 5, 500)
plt.plot(x, -(x**3) + 2 * x + np.exp(x) - 1)
plt.plot(x, x**2 - 1, color='green')
plt.grid(True, linestyle='-', color='0.75')
plt.show()
```

```python
def f1(p):
    x, y = p
    return (y - x**2 + 1, x*(-y) + x + np.exp(x) - 1)

x11, y11 = fsolve(f1, (-2, 3))
x21, y21 = fsolve(f1, (-0.3, -0.2))
x31, y31 = fsolve(f1, (1.9, 2.5))
x41, y41 = fsolve(f1, (5, 24))
print(x11, y11)
print(x21, y21)
print(x31, y31)
print(x41, y41)
```

```python
plt.figure(figsize=(5, 5))
x = np.linspace(-4, 5, 500)
plt.plot(x, -(x ** 3) + 2 * x + np.exp(x) - 1)
plt.plot(x, x**2 - 1, color='green')
plt.grid(True, linestyle='-', color='0.75')
plt.scatter(x11, y11, c = 'deeppink')
plt.scatter(x21, y21, c = 'deeppink')
plt.scatter(x31, y31, c = 'deeppink')
plt.scatter(x41, y41, c = 'deeppink')
plt.show()
```

т.е. здесь появляются диапазоны решений:

x < -1.581835 => y = x^2 - 1 = 1.502203

0 < x < 2.618145 => y = x^2 - 1 = 5.854686

x > 4.200105 => y = x^ - 1 = 16.640889
