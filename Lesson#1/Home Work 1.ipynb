```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(21)
k = 1
k1 = np.random.randint(2, 10)
print(f'k1={k1}')
x = np.linspace(-20, 20, 1000)
y = np.cos(k * x)
y1 = np.cos(k1 * x)
```

```python
plt.figure(figsize=(15,5))
plt.plot(x, y)
plt.plot(x, y1)
plt.show
```
